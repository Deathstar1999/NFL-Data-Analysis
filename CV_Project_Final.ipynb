```python
import pandas as pd # import pandas 
Basic_Stats=pd.read_csv('Basic_Stats.csv') # reading the files
Game_Logs_Wide_Receiver_and_Tight_End=pd.read_csv('Game_Logs_Wide_Receiver_and_Tight_End.csv',na_values=["--"])
Career_Stats_Fumbles=pd.read_csv('Career_Stats_Fumbles.csv',na_values=["--"])
Game_Logs_Defensive_Lineman=pd.read_csv('Game_Logs_Defensive_Lineman.csv',na_values=["--"])
Game_Logs_Quarterback=pd.read_csv('Game_Logs_Quarterback.csv',na_values=["--"])
Game_Logs_Runningback=pd.read_csv('Game_Logs_Runningback.csv',na_values=["--"])
Game_Logs_Kickers=pd.read_csv('Game_Logs_Kickers.csv',na_values=["--"])
Career_Stats_Kick_Return=pd.read_csv('Career_Stats_Kick_Return.csv',na_values=["--"])
Career_Stats_Defensive=pd.read_csv('Career_Stats_Defensive.csv',na_values=["--"])
Career_Stats_Receiving=pd.read_csv('Career_Stats_Receiving.csv',na_values=["--"])
Game_Logs_Offensive_Line=pd.read_csv('Game_Logs_Offensive_Line.csv',na_values=["--"])
Game_Logs_Punters=pd.read_csv('Game_Logs_Punters.csv',na_values=["--"])
Career_Stats_Rushing=pd.read_csv('Career_Stats_Rushing.csv',na_values=["--"])

# filtering the recent years from the datasets and sorting the datasets in descending order
Game_Logs_Offensive_Line=Game_Logs_Offensive_Line[(Game_Logs_Offensive_Line["Year"]>2010) & 
                         (Game_Logs_Offensive_Line["Season"]=="Regular Season")].sort_values(by="Year",ascending=False)
Game_Logs_Wide_Receiver_and_Tight_End=Game_Logs_Wide_Receiver_and_Tight_End[(Game_Logs_Wide_Receiver_and_Tight_End["Year"]>2010) & 
                                      (Game_Logs_Wide_Receiver_and_Tight_End["Season"]=="Regular Season")].sort_values(by="Year",ascending=False)
Game_Logs_Defensive_Lineman=Game_Logs_Defensive_Lineman.loc[(Game_Logs_Defensive_Lineman["Year"]>2010)
                                & (Game_Logs_Defensive_Lineman["Season"]=="Regular Season")].sort_values(by="Year",ascending=False)

Game_Logs_Quarterback=Game_Logs_Quarterback.loc[(Game_Logs_Quarterback["Year"]>2010)
                          & (Game_Logs_Quarterback["Season"]=="Regular Season")].sort_values(by="Year",ascending=False)
Game_Logs_Runningback=Game_Logs_Runningback.loc[(Game_Logs_Runningback["Year"]>2010)
                          & (Game_Logs_Runningback["Season"]=="Regular Season")].sort_values(by="Year",ascending=False)
Game_Logs_Kickers=Game_Logs_Kickers.loc[(Game_Logs_Kickers["Year"]>2010) 
                      & (Game_Logs_Kickers["Season"]=="Regular Season")].sort_values(by="Year",ascending=False)
Game_Logs_Punters=Game_Logs_Punters.loc[(Game_Logs_Punters["Year"]>2010) 
                      & (Game_Logs_Punters["Season"]=="Regular Season")].sort_values(by="Year",ascending=False)
Career_Stats_Fumbles=Career_Stats_Fumbles.loc[(Career_Stats_Fumbles["Year"]>2010)].sort_values(by="Year",ascending=False)
Career_Stats_Kick_Return=Career_Stats_Kick_Return.loc[(Career_Stats_Kick_Return["Year"]>2010)].sort_values(by="Year",ascending=False)
Career_Stats_Defensive=Career_Stats_Defensive.loc[(Career_Stats_Defensive["Year"]>2010)].sort_values(by="Year",ascending=False)
Career_Stats_Receiving=Career_Stats_Receiving.loc[(Career_Stats_Receiving["Year"]>2010)].sort_values(by="Year",ascending=False)
Career_Stats_Rushing=Career_Stats_Rushing.loc[(Career_Stats_Rushing["Year"]>2010)].sort_values(by="Year",ascending=False)

# extracting the active players from the Basics stats file
Active_players=list(Basic_Stats[Basic_Stats["Current Status"].isin(["Active","Unsigned free agent","Injured reserve"])]["Name"].unique())

# filtering the active players only from all the stats files
Career_Stats_Receiving=Career_Stats_Receiving[Career_Stats_Receiving['Name'].isin(Active_players)]                                
Game_Logs_Offensive_Line=Game_Logs_Offensive_Line[Game_Logs_Offensive_Line['Name'].isin(Active_players)]
Game_Logs_Wide_Receiver_and_Tight_End=Game_Logs_Wide_Receiver_and_Tight_End[Game_Logs_Wide_Receiver_and_Tight_End['Name'].isin(Active_players)]
Game_Logs_Defensive_Lineman=Game_Logs_Defensive_Lineman[Game_Logs_Defensive_Lineman['Name'].isin(Active_players)]
Game_Logs_Quarterback=Game_Logs_Quarterback[Game_Logs_Quarterback['Name'].isin(Active_players)]
Game_Logs_Runningback=Game_Logs_Runningback[Game_Logs_Runningback['Name'].isin(Active_players)]
Game_Logs_Kickers=Game_Logs_Kickers[Game_Logs_Kickers['Name'].isin(Active_players)]
Game_Logs_Punters=Game_Logs_Punters[Game_Logs_Punters['Name'].isin(Active_players)]
Career_Stats_Fumbles=Career_Stats_Fumbles[Career_Stats_Fumbles['Name'].isin(Active_players)]
Career_Stats_Kick_Return=Career_Stats_Kick_Return[Career_Stats_Kick_Return['Name'].isin(Active_players)]
Career_Stats_Defensive=Career_Stats_Defensive[Career_Stats_Defensive['Name'].isin(Active_players)]
Career_Stats_Rushing=Career_Stats_Rushing[Career_Stats_Rushing['Name'].isin(Active_players)]


# filling the empty positions where the values are Nan with "--"
Game_Logs_Wide_Receiver_and_Tight_End["Position"].fillna("--",inplace=True)
Career_Stats_Fumbles["Position"].fillna("--",inplace=True)
Game_Logs_Defensive_Lineman["Position"].fillna("--",inplace=True)
Game_Logs_Quarterback["Position"].fillna("--",inplace=True)
Game_Logs_Runningback["Position"].fillna("--",inplace=True)
Game_Logs_Kickers["Position"].fillna("--",inplace=True)
Career_Stats_Kick_Return["Position"].fillna("--",inplace=True)
Career_Stats_Defensive["Position"].fillna("--",inplace=True)
Career_Stats_Receiving["Position"].fillna("--",inplace=True)
Game_Logs_Offensive_Line["Position"].fillna("--",inplace=True)
Game_Logs_Punters["Position"].fillna("--",inplace=True)
Career_Stats_Rushing["Position"].fillna("--",inplace=True)


# Filling all the rest of the stats columns with 0s
Game_Logs_Wide_Receiver_and_Tight_End.fillna(0,inplace=True)
Career_Stats_Fumbles.fillna(0,inplace=True)
Game_Logs_Defensive_Lineman.fillna(0,inplace=True)
Game_Logs_Quarterback.fillna(0,inplace=True)
Game_Logs_Runningback.fillna(0,inplace=True)
Game_Logs_Kickers.fillna(0,inplace=True)
Career_Stats_Kick_Return.fillna(0,inplace=True)
Career_Stats_Defensive.fillna(0,inplace=True)
Career_Stats_Receiving.fillna(0,inplace=True)
Game_Logs_Offensive_Line.fillna(0,inplace=True)
Game_Logs_Punters.fillna(0,inplace=True)
Career_Stats_Rushing.fillna(0,inplace=True)


# bringing the year values to the recent values for visualisation
Game_Logs_Offensive_Line["Year"]=Game_Logs_Offensive_Line["Year"]+6
Game_Logs_Wide_Receiver_and_Tight_End["Year"]=Game_Logs_Wide_Receiver_and_Tight_End["Year"]+6
Game_Logs_Defensive_Lineman["Year"]=Game_Logs_Defensive_Lineman["Year"]+6
Game_Logs_Quarterback["Year"]=Game_Logs_Quarterback["Year"]+6
Game_Logs_Runningback["Year"]=Game_Logs_Runningback["Year"]+6
Game_Logs_Kickers["Year"]=Game_Logs_Kickers["Year"]+6
Game_Logs_Punters["Year"]=Game_Logs_Punters["Year"]+6
Career_Stats_Fumbles["Year"]=Career_Stats_Fumbles["Year"]+6
Career_Stats_Kick_Return["Year"]=Career_Stats_Kick_Return["Year"]+6
Career_Stats_Defensive["Year"]=Career_Stats_Defensive["Year"]+6
Career_Stats_Receiving["Year"]=Career_Stats_Receiving["Year"]+6
Career_Stats_Rushing["Year"]=Career_Stats_Rushing["Year"]+6

# function to replace column values and convert columns to float data type
def convert_to_float(entry):
    if type(entry)==str and entry[len(entry)-1]=="T":
        return float(entry[0:len(entry)-1])
    else:
        return float(entry)

# dropping some irrelevant columns
Career_Stats_Kick_Return.drop("Yards Returned",axis=1,inplace=True)
Career_Stats_Receiving.drop("Receiving Yards",axis=1,inplace=True)
Career_Stats_Rushing.drop("Rushing Yards",axis=1,inplace=True)

# applying the convert to float function to the relevant columns
Career_Stats_Rushing["Longest Rushing Run"]=Career_Stats_Rushing["Longest Rushing Run"].apply(convert_to_float)
Career_Stats_Kick_Return["Longest Return"]=Career_Stats_Kick_Return["Longest Return"].apply(convert_to_float)
Career_Stats_Defensive["Longest Int Return"]=Career_Stats_Defensive["Longest Int Return"].apply(convert_to_float)
Career_Stats_Receiving["Longest Reception"]=Career_Stats_Receiving["Longest Reception"].apply(convert_to_float)
Game_Logs_Defensive_Lineman["Longest Int Return"]=Game_Logs_Defensive_Lineman["Longest Int Return"].apply(convert_to_float)
Game_Logs_Runningback["Longest Reception"]=Game_Logs_Runningback["Longest Reception"].apply(convert_to_float)
Game_Logs_Runningback["Longest Rushing Run"]=Game_Logs_Runningback["Longest Rushing Run"].apply(convert_to_float)
Game_Logs_Wide_Receiver_and_Tight_End["Longest Reception"]=Game_Logs_Wide_Receiver_and_Tight_End["Longest Reception"].apply(convert_to_float)
Game_Logs_Wide_Receiver_and_Tight_End["Longest Rushing Run"]=Game_Logs_Wide_Receiver_and_Tight_End["Longest Rushing Run"].apply(convert_to_float)
```
